## What is the distribution of sales (discounted price vs. actual price)?

```python
# Function to remove the currency symbol and convert to numeric
def remove_currency_symbol_and_convert_to_numeric(value):
    return pd.to_numeric(value.replace('₹', ''), errors='coerce')

# Convert price columns to numeric format
df['discounted_price'] = df['discounted_price'].apply(remove_currency_symbol_and_convert_to_numeric)
df['actual_price'] = df['actual_price'].apply(remove_currency_symbol_and_convert_to_numeric)

# Calculate the discount percentage for each product
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# Create bins for discount percentage ranges
bins = [0, 25, 50, 75, 100]

# Labels for each bin
labels = ['0-24%', '25-49%', '50-74%', '75-100%']

# Categorize products into discount percentage ranges
df['discount_range'] = pd.cut(df['discount_percentage'], bins=bins, labels=labels, include_lowest=True)

# Count the number of products in each discount percentage range
sales_distribution = df['discount_range'].value_counts()

# Plot the distribution using a bar chart
plt.figure(figsize=(8, 6))
plt.bar(sales_distribution.index, sales_distribution.values, color='skyblue')
plt.xlabel('Discount Percentage Range')
plt.ylabel('Number of Products')
plt.title('Distribution of Sales (Discounted Price vs. Actual Price)')
plt.xticks(rotation=45)
plt.show()
```

## What is the distribution of sales by product category?

## What is the average rating and rating count for products?

```python
# Convert the 'rating' column to numeric format, handling any non-numeric values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
# Calculate the average rating
average_rating = df['rating'].mean()

# Calculate the total rating count for all products
total_rating_count = df['rating_count'].sum()

# Calculate the average rating count per product
average_rating_count_per_product = total_rating_count / len(df)

print(f"Average Rating: {average_rating:.2f}")
print(f"Average Rating Count per Product: {average_rating_count_per_product:.2f}")
```

## What is the distribution of products across different categories?

```python
# Split the sub-categories from the 'category' column
df['category'] = df['category'].apply(lambda x: x.split('|')[0])

# Count the number of products in each category
product_count_by_category = df['category'].value_counts()

# Plot the distribution using a bar chart
plt.figure(figsize=(10, 6))
product_count_by_category.plot(kind='bar', color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Number of Products')
plt.title('Distribution of Products Across Different Categories')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## What is the correlation between ratings, rating count, and discounted price?
